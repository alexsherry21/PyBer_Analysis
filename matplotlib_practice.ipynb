```python
%matplotlib inline

#import dependencies
import matplotlib.pyplot as plt
```

```python
#Set the x-axis equal to a list of strings for each mont
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

#Set the y-axis equal to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
#plt.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plot()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plot()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="orange", linewidth=2, label = "Boston")
#create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set the y limit between 0 and 45
plt.ylim(0,45)
#create a title
plt.title("PyBer Fare by Month")
#add the legend
plt.legend()
#add a grid
plt.grid()
```

```python
#object-oriented skill drill
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="purple", linewidth=2, label="Boston")
ax.set_ylim(0,45)
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.grid()
ax.legend()
```

```python
#create bar chart
plt.bar(x_axis, y_axis, color="red", label="Boston")
#create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#create a title
plt.title("PyBer Fare by Month")
#add a legend
plt.legend()
```

```python
#create horizontal chart
plt.barh(x_axis, y_axis, label = "Boston", color = "magenta")
plt.gca().invert_yaxis()
plt.legend()
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare ($)")
```

```python
#create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
#create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label = "Chicago")
ax.invert_yaxis()
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.legend()
ax.set_title("PyBer Fare by Month")
```

```python
plt.plot(x_axis, y_axis, "o")
```

```python
plt.scatter(y_axis, x_axis, color="green", label="Chicago")
plt.xlim(0,45)
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis, color="grey")
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3.5)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger, color="teal")
```

```python
plt.scatter(x_axis, y_axis, s= [i * 3.5 for i in y_axis], color="brown")
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color = "skyblue", alpha=0.8, s=[i*5 for i in y_axis], label = "Boston", linewidths=2, edgecolors="black")
ax.invert_yaxis()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.legend()
ax.set_xlim(0,50)
```

```python
explode_values=(0, 0, .2, 0, 0, 0, .2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(9,9))
plt.pie(y_axis, labels=x_axis, autopct="%.1f%%", explode=explode_values, colors=colors)
```
